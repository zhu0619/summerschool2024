## For ML, molecules need to be converted into a list of features/representation.

In this tutorial, we will focus on **Fingerprint** and **Molecular Graph**.

### Morgan fingerprints ECFP
Extended-Connectivity Fingerprints (ECFPs) are circular topological fingerprints designed for molecular characterization, similarity searching, and structure-activity modeling. (see ChemAxons ECFP docs) The original implementation of the ECFPs was done in Pipeline Pilot which is not open-source. Instead we use the implementation from RDKit which is called Morgan fingerprint. The two most important parameters of these fingerprints are the radius and fingerprint length. The first specifies the radius of circular neighborhoods considered for each atom. Here two radii are considered: 2 and 3. The length parameter specifies the length to which the bit string representation is hashed. The default length is 2048.

### Use `datamol` to quickly compute ecfp4 fingerprints|

###  Molecular graph features

### Process smiles into `torch_geometric.Data` objects

```python
from molfeat.trans.graph.adj import PYGGraphTransformer

# convert smiles to Rdkit.Mol object
mols = dm.parallelized(fn=dm.to_mol, inputs_list=egfr[mol_col], progress=True)

# conver Rdkit.Mol object to `torch_geometric.Data` object
transformer = PYGGraphTransformer()
egfr["pyg"] = transformer(mols)
```

#### Molecular graph example

```python
pyg_mol = egfr.loc[0, "pyg"]
pyg_mol
```

Visualize the adjacent matrix of the graph

```python
from torch_geometric.utils import to_dense_adj

# convert the graph object to adjacent matrix for visualization
to_dense_adj(pyg_mol.edge_index)
```

```python
from torch_geometric.utils import to_networkx
from matplotlib import pyplot as plt
import networkx as nx

G = to_networkx(pyg_mol, to_undirected=True)
plt.figure(figsize=(12,12))
plt.axis('off')
nx.draw_networkx(G,
                pos=nx.spring_layout(G, seed=0),
                with_labels=True,
                node_size=800,
                cmap="hsv",
                vmin=-2,
                vmax=3,
                width=0.8,
                edge_color="grey",
                font_size=14
                )
plt.show()
```
